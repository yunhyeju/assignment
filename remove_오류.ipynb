```python
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

class LinkedList:

    # 초기화 메소드
    def __init__(self):
        dummy = Node("dummy")
        self.head = dummy
        self.tail = dummy

        self.current = None
        self.before = None

        self.num_of_data = 0

    # append 메소드 (insert - 맨 뒤에 노드 추가, tail과 node의 next, 데이터 개수 변경)
    def append(self, data):
        new_node = Node(data)
        self.tail.next = new_node
        self.tail = new_node

        self.num_of_data += 1

    # delete 메소드 (delete - current 노드 삭제, 인접 노드의 current, next 변경, 데이터 개수 변경)
    def delete(self):
        pop_data = self.current.data

        if self.current is self.tail:
            self.tail = self.before

            # 중요 : current가 next가 아닌 before로 변경된다.
            self.before.next = self.current.next
            self.current = self.before 

            self.num_of_data -= 1

            return pop_data

    # first 메소드 (search1 - 맨 앞의 노드 검색, before, current 변경)
    def first(self):
        # 데이터가 없는 경우 첫번째 노드도 없기 때문에 None 리턴
        if self.num_of_data == 0: 
            return None

        self.before = self.head
        self.current = self.head.next

        return self.current.data

    # next 메소드 (search2 - current 노드의 다음 노드 검색, 이전에 first 메소드가 한번은 실행되어야 함)
    def next(self):
        if self.current.next == None:
            return None

        self.before = self.current
        self.current = self.current.next

        return self.current.data

    # size 메소드
    def size(self):
        return self.num_of_data
    
    def traverse_all(self):
        print("head",end ='->')
        data = l_list.first()
        if data:
            print("(%s)"%(data), end= '->')
        while True:
            data = l_list.next()
            if data:
                print("(%s)"%(data),end="->")
            elif data == None:
                print("null")
                break
 
        
        
    def insert_at(self,position,new_data):
        new_node = Node(new_data)
        #self.current = self.head
        

        if position < 0 : # index가 0보다 작다면
            print("error") # error 출력
                
        elif position > self.num_of_data :  # 원래의 노드 넘버보다 큰 position 값이 들어온다면
            self.tail.next = new_node #tail, 즉 원래의 노드 제일 끝 다음에 new node를 insert
            self.tail = new_node #tail은 new node가 됨
            self.num_of_data += 1 #노드 넘버에 추가된 노드 반영
            
        elif position == 1: # position이 1인경우

            new_node.next = self.head.next
            self.head.next = new_node 
            
        else:
            self.current = self.head # 현재 노드를 head로 설정정
            for i in range(position-2):#position이 3이라면 2번 움직이도록
                self.current = self.current.next 
            new_node.next = self.current.next
            self.current.next = new_node
        self.num_of_data += 1
        
                
    def remove(self, key):
        self.current = self.head
        self.before = None
        removed = False
        while self.current is not None:
            if self.current.data == key:
                if self.before is None:
                    self.head = self.current.next #첫 번째 노드(head)가 key인 경우, 첫 번째 노드는 삭제 되고, 그 다음 노드가 head가 됨
                    self.current = self.head #current node로 된 것이 head 가 됨
                else: #그렇지 않은 경우
                    self.before.next = self.current.next # 이전 노드의 다음이, 현재 노드의 다음이 되고
                    self.current = self.current.next #현재 노드는, 원래 현재 노드의 다음이 됨
                removed = True
            else: #current node와 키가 같지 않으면 그냥 옮겨감.
                self.current = self.current.next 


        if removed:
            print(f"{key} 값을 가지는 노드를 모두 삭제했습니다.")
        else:
            print(f"해당하는 원소가 없습니다.")

                
                
            
            
        
        
      
            
            
            
            

            
                
                
            
      
            
            
        
```
