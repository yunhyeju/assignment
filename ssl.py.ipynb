## remove
+ 현재 노드가 none이 아닌 경우 -> key 값과 current node 값이 같은 경우 -> current node의 이전 노드가 없는 경우(current node가 head 인 경우)  : 첫 번째 노드는 삭제되고 첫 번째 노드의 다음노드가 head가 됨. )
+ 현재 노드가 none이 아닌 경우 -> key 값과 current node 값이 같은 경우 -> current node의 이전 노드가 있는 경우(current node가 중간 노드인 경우) : # 이전 노드의 다음이, 현재 노드의 다음이 되고 , 현재 노드는, 현재 노드의 다음이 됨. (즉, 현재 노드가 removed됨)
+ 위 작업을 진행하면 removed = True 가 됨

+ 현재 노드가 none이 아닌 경우 -> key 값과 current node 값이 다른 경우 : 그냥 옮겨감. 이전 노드와 현재 노드 관계 설정, 현재 노드와 현재 다음노드의 관계를 설정하는 것에서 그침.

+ removed = True가 되면 f"{key} 값을 가지는 노드를 모두 삭제했습니다." 출력, 일치하는 키 값이 없으면 f"해당하는 원소가 없습니다." 출력.

```python

class LinkedList:
    def __init__(self):
        dummy = Node("dummy")
        self.head = dummy
        self.tail = dummy

        self.current = None
        self.before = None

        self.num_of_data = 0
  

    def append(self,data):
        new_node = Node(data)
        self.tail.next = new_node #꼬리 끝에 새 노드 추가 (꼬리 다음 = 새 노드)
        self.tail = new_node # 새 노드가 꼬리가 됨
        self.num_of_data +=1


    def delete(self):
        pop_data = self.current.data

        if self.current is self.tail:
            self.tail = self.before # 꼬리 이전의 노드가 꼬리가 됨

            self.before.next = self.current.next #이전의 다음 노드 = 현재의 다음 노드가 됨
            self.current = self.before
            self.num_of_data -=1
            return pop_data


    def first(self):
        if self.num_of_data == 0:
            return None

        self.before = self.head # current 이전의 노드는 head node 즉 dummy
        self.current = self.head.next # current 는 dummy 바로 다음의 노드가 됨.즉 값이 제일 처음 들어있는 노드가 current 노드가 됨
        return self.current.data # 해당 값 반환

    def next(self): # current 노드의 다음 노드 검색, 이전에 first 메소드가 한번은 실행되어야 함 first에서 정확한 노드를 before, current로 잡고있기 때문문
        if self.current.next == None:
            return None
    
        self.before = self.current # 원래 현재 이전의 노드를 현재 노드로 보고
        self.current = self.current.next # 현재 노드는, 원래 현재의 다음 노드가 됨.

        return self.current.data

    def size(self):
        return self.num_of_data


    def traverse_all(self):
        print("head",end ='->')
        data = l_list.first() # tranvers_all 메소드에서는 first라는 메소드를 씀 / 즉  값이 제일 처음 들어있는 노드 값이 data 변수에 들어감
        if data:
            print("(%s)"%(data), end= '->')
        while True:
            data = l_list.next() # next 메소드도 사용함.
            if data :
                print("(%s)"%(data),end="->")
            else:
                print("null")
                break

    def insert_at(self,position,new_data):
        new_node = Node(new_data)
        if position < 0:
            print("error")

        elif position > self.num_of_data:
            self.tail.next = new_node
            self.tail = new_node

        elif position == 1:
            new_node.next = self.head.next #dummy(head) -> new_node -> head.next(new_node.next)
            self.head.next = new_node #dummy(head) -> new_node(head.next)

        else:
            self.current = self.head # 현재 노드를 head로 설정정
            for i in range(position-2):#position이 3이라면 2번 움직이도록
                self.current = self.current.next 
            new_node.next = self.current.next
            self.current.next = new_node
        self.num_of_data += 1


    def remove(self,key):
        self.current = self.head 
        self.before = None
        removed = False

        while self.current != None:
            if self.current.data == key:
                if self.before is None:
                    self.head = self.current.next #첫 번째 노드(head)가 key인 경우, 첫 번째 노드는 삭제 되고, 그 다음 노드가 head가 됨
                    self.current = self.current.next #current node로 된 것이 head 가 됨
                else: #그렇지 않은 경우
                    self.before.next = self.current.next # 이전 노드의 다음이, 현재 노드의 다음이 되고
                    self.current = self.current.next #현재 노드는, 원래 현재 노드의 다음이 됨
                removed = True
            else: # current node와 키가 같지 않으면 그냥 옮겨감.
                self.before = self.current # 이전 노드와 현재 노드의 관계설정
                self.current = self.current.next #현재 노드와 다음 노드의 관계 설정

        if removed:
            print(f"{key} 값을 가지는 노드를 모두 삭제했습니다.")
        else:
            print(f"해당하는 원소가 없습니다.")

```
